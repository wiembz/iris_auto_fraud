## Pourquoi SHAP n'est pas utilisé pour VHS

Le VHS n'est pas expliqué par SHAP car il ne s'agit pas d'un modèle prédictif de type machine learning. Le score est calculé à partir de règles explicites et de points de contrôle STAFFIM observables. L'explication retenue est donc une explication déterministe par contributions de checkpoints, plus adaptée à un usage métier assurantiel.

SHAP est utile pour expliquer des modèles prédictifs de machine learning. Le VHS ne prédit pas une probabilité de fraude, ne calcule pas un risque statistique de sinistre et ne repose pas sur un modèle opaque. Il applique des règles métier déterministes : observation initiale, lecture standardisée, niveau de sensibilité, impact sur l'état du véhicule et limitation métier éventuelle du niveau final.

Chaque réduction de l'indicateur peut être rattachée directement à un point de contrôle STAFFIM observable, une observation initiale, une lecture métier standardisée et un impact explicite sur l'état du véhicule.

Cette approche améliore la transparence pour les utilisateurs de BNA Assurances : chaque raison affichée est observable, vérifiable et auditable.


```python
# -- Business-readable VHS label mapping

HUMAN_IN_THE_LOOP_DISCLAIMER = (
    "Les éléments présentés constituent une aide à l'analyse. "
    "La décision finale reste sous la responsabilité du gestionnaire."
)

decision_labels = {
    'OK': 'État satisfaisant',
    'DEGRADE': 'État à surveiller',
    'IMMOBILISE': 'Usage déconseillé',
    'CRITIQUE': 'Examen prioritaire suggéré',
}

status_labels = {
    'OK': 'Élément conforme',
    'WORN': 'Usure observée',
    'WORN_STRONG': 'Intervention conseillée',
    'BROKEN': 'Défaut confirmé',
    'REPAIRED': 'Élément réparé',
    'UNKNOWN': 'Information non exploitable',
}

grade_labels = {
    'A': 'Aucun signal technique majeur',
    'B': 'Quelques points à surveiller',
    'C': 'Dégradation technique notable',
    'D': 'Situation technique sensible',
}

technical_term_replacements = {
    'score': "Indicateur d'état",
    'grade': "Niveau d'état technique",
    'checkpoint': 'Point de contrôle',
    'penalty': "Impact sur l'état du véhicule",
    'règle de limitation du niveau final': 'Limitation métier du niveau final',
    'normalized status': "Lecture standardisée de l'observation",
    'is_immobilizing': "Point sensible pour l'usage du véhicule",
    'WORN_STRONG': 'Intervention conseillée',
    'BROKEN': 'Défaut confirmé',
}

def business_decision_label(code):
    return decision_labels.get(str(code), 'Libellé métier non disponible')

def business_status_label(code):
    return status_labels.get(str(code), 'Lecture métier non disponible')

def business_grade_label(code):
    code = str(code)
    label = grade_labels.get(code, 'Niveau non disponible')
    return f'{code} — {label}' if code in grade_labels else label

def _safe_num(value, default=0.0):
    try:
        if value is None:
            return default
        return float(value)
    except Exception:
        return default

def impact_label(value):
    val = _safe_num(value)
    if val <= 0:
        return 'Aucun'
    if val >= 20:
        return 'Fort'
    if val >= 10:
        return 'Moyen'
    return 'Faible'

def sensitivity_label(tier):
    tier = str(tier or '').upper()
    if 'T1' in tier or 'VITAL' in tier:
        return 'Sécurité prioritaire'
    if 'T2' in tier or 'CRITICAL' in tier:
        return 'Fonctionnement important'
    if 'T3' in tier or 'COSMETIC' in tier:
        return 'Confort ou apparence'
    return 'Sensibilité non précisée'

def _first_present(row, names, default='N/A'):
    for name in names:
        try:
            value = row.get(name)
        except Exception:
            value = None
        if value is not None and str(value).strip() not in ('', 'nan', 'None'):
            return value
    return default

def checkpoint_business_name(row):
    return _first_present(row, ['checkpoint_libelle', 'checkpoint_label', 'checkpoint_code'], 'Point de contrôle non renseigné')

def observation_initiale(row):
    return _first_present(row, ['valeur_controle', 'observed_value', 'observation_initiale'], 'Observation non renseignée')

def technical_status(row):
    return _first_present(row, ['normalized_status', 'observed_status'], 'UNKNOWN')

def build_business_reason_table(penalties, top_n=5):
    if penalties is None or len(penalties) == 0:
        return pd.DataFrame(columns=['point_de_controle', 'observation_initiale', 'lecture_metier', 'niveau_sensibilite', 'impact_etat_vehicule', 'decision_concernee'])
    work = penalties.copy()
    if 'penalty_applied' in work.columns:
        work = work.sort_values('penalty_applied', ascending=False)
    rows = []
    for _, p in work.head(top_n).iterrows():
        status = technical_status(p)
        impact_value = p.get('penalty_applied', p.get('total_penalty', 0))
        decision_code = p.get('decision', '')
        rows.append({
            'point_de_controle': checkpoint_business_name(p),
            'observation_initiale': observation_initiale(p),
            'lecture_metier': business_status_label(status),
            'niveau_sensibilite': sensitivity_label(p.get('tier', '')),
            'impact_etat_vehicule': impact_label(impact_value),
            'decision_concernee': business_decision_label(decision_code) if decision_code else 'Voir résultat inspection',
            'technical_status': status,
            'impact_numerique': round(_safe_num(impact_value), 2),
        })
    return pd.DataFrame(rows)

def format_score(value):
    try:
        return f'{float(value):.1f}'
    except Exception:
        return str(value if value is not None else 'N/A')

def print_business_case(row, penalties=None, label=''):
    if label:
        print(f'\n  {label}')
        print('  ' + '─' * 58)
    inspection_id = row.get('inspection_key', row.get('inspection_id', 'N/A'))
    score = format_score(row.get('vhs_final_score', row.get('score', 'N/A')))
    grade = row.get('safety_grade', row.get('grade', 'N/A'))
    decision = row.get('decision', 'N/A')
    print('  A. Résumé technique')
    print(f'     inspection_id : {inspection_id}')
    print(f'     score         : {score}')
    print(f'     grade         : {grade}')
    print(f'     decision      : {decision}')
    print('\n  B. Résumé métier')
    print(f"     Indicateur d'état : {score} / 100")
    print(f"     Niveau d'état technique : {business_grade_label(grade)}")
    print(f"     Niveau d'attention proposé : {business_decision_label(decision)}")
    reasons = build_business_reason_table(penalties, top_n=5)
    print('\n  C. Principales raisons observables')
    if len(reasons) == 0:
        print('     Aucun impact majeur détaillé disponible dans le contexte chargé.')
    else:
        cols = ['point_de_controle', 'observation_initiale', 'lecture_metier', 'impact_etat_vehicule']
        print(reasons[cols].to_string(index=False))
    print(f'\n  D. {HUMAN_IN_THE_LOOP_DISCLAIMER}')

print('[OK] Libellés métier VHS définis pour les explications non techniques.')
```

```python
# -- Validation helper functions

def section_title(title):
    sep = '=' * 72
    print(f'\n{sep}\n  {title}\n{sep}')

def pass_fail(condition, warn=False):
    if condition:
        return 'PASS'
    return 'WARNING' if warn else 'FAIL'

def pct(part, total):
    return '0.0%' if total == 0 else f'{100 * part / total:.1f}%'

def safe_read_sql(engine, query, params=None):
    '''Execute SELECT query; return empty DataFrame on error.'''
    try:
        from sqlalchemy import text
        with engine.connect() as conn:
            if params:
                return pd.read_sql(text(query), conn, params=params)
            return pd.read_sql(text(query), conn)
    except Exception as exc:
        print(f'  [DB WARNING] {type(exc).__name__}: {str(exc)[:120]}')
        return pd.DataFrame()

def display_check(name, expected, actual, ok=None):
    if ok is None:
        ok = (expected == actual)
    marker = '[PASS]' if ok else '[FAIL]'
    print(f'  {marker}  {name}: expected={expected}, actual={actual}')
    return ok

print('[OK] Helper functions defined.')
```

## Section 5 — Distribution des décisions

Le VHS produit quatre décisions possibles pour chaque véhicule inspecté :

## Section 6 — Distribution des notes de sécurité

Parallèlement à la décision, chaque inspection reçoit une **note de sécurité** (A, B, C, D) basée sur l'état des checkpoints vitaux et importants :

| Note | Signification |
|------|---------------|
| **A** | Aucun checkpoint vital ou important défaillant |
| **B** | Checkpoints T1 WORN ou WORN_STRONG |
| **C** | Checkpoint vital WORN_STRONG ou important BROKEN |

## Section 7 — Traitement des valeurs ambiguës

Cette section est **critique** pour la validation de la version V3.

Dans la version V2, les valeurs `PROPOSITION FAITE` et `NON` tombaient dans le cas générique `est_anomalie_critique = true → BROKEN`, ce qui conduisait à une sur-pénalisation de cas ambigus.

**Règles V3 pour les valeurs ambiguës :**

| Valeur observée | Condition | Mapping V3 |
|-----------------|-----------|------------|
| `PROPOSITION FAITE` | anomalie critique | `WORN_STRONG` |
| `PROPOSITION FAITE` | anomalie simple | `WORN` |
| `NON` | anomalie critique | `WORN_STRONG` |
| `NON` | anomalie simple | `WORN` |

**Contrat V3 :** Ces valeurs ne doivent jamais être mappées à `BROKEN`. Toute occurrence constitue une violation de contrat.


```python
section_title('Section 7 — Traitement des valeurs ambigues')

def norm_val(v):
    if v is None or (isinstance(v, float) and np.isnan(v)):
        return ''
    import unicodedata
    nfd = unicodedata.normalize('NFD', str(v).strip().lower())
    return ' '.join(c for c in nfd if unicodedata.category(c) != 'Mn').split()
    # Simplified: just normalize spaces and lowercase

if len(df_penalty) > 0 and 'observed_value' in df_penalty.columns:
    # Check PROPOSITION FAITE → BROKEN
    mask_prop = df_penalty['observed_value'].astype(str).str.lower().str.strip().str.contains(
        'proposition faite', na=False
    )
    prop_broken = int(
        (mask_prop & (df_penalty['observed_status'] == 'BROKEN')).sum()
    )
    # Check NON → BROKEN
    mask_non = df_penalty['observed_value'].astype(str).str.lower().str.strip() == 'non'
    non_broken = int(
        (mask_non & (df_penalty['observed_status'] == 'BROKEN')).sum()
    )
    # Total ambiguous rows
    total_prop = int(mask_prop.sum())
    total_non  = int(mask_non.sum())
else:
    prop_broken = DOCUMENTED['proposition_to_broken']
    non_broken  = DOCUMENTED['non_to_broken']
    total_prop  = None
    total_non   = None
    print('[INFO] Verifications sur valeurs documentees.')

print()
print('  -- Contrat V3: valeurs ambigues ne doivent PAS mapper vers BROKEN --')
print()
display_check('PROPOSITION FAITE -> BROKEN = 0', 0, prop_broken)
display_check('NON -> BROKEN = 0',               0, non_broken)
print()
if total_prop is not None:
    print(f'  Total lignes PROPOSITION FAITE : {total_prop}')
    print(f'  Total lignes NON               : {total_non}')
print()
if prop_broken == 0 and non_broken == 0:
    print('  RESULTAT: Contrat V3 RESPECTE.')
    print('  Les valeurs ambigues (PROPOSITION FAITE, NON) sont correctement traitees')
    print('  comme WORN ou WORN_STRONG, jamais comme BROKEN.')
else:
    print('  RESULTAT: VIOLATION DU CONTRAT V3 DETECTEE.')
    print(f'  PROPOSITION FAITE -> BROKEN: {prop_broken}')
    print(f'  NON -> BROKEN: {non_broken}')
```

```python
section_title('Section 14b — Validation par exemples multiples de données réelles')
print('B. Chargement des fonctions utilitaires orientées métier')
print()

def mask_identifier(val):
    if val is None:
        return 'N/A'
    s = str(val)
    return (s[:-4] + '****') if len(s) > 4 else '****'

def get_case_penalties(df_pen, inspection_key, run_id=None):
    if df_pen is None or len(df_pen) == 0:
        return pd.DataFrame()
    mask = df_pen['inspection_key'] == inspection_key
    if run_id is not None and 'run_id' in df_pen.columns:
        mask = mask & (df_pen['run_id'] == run_id)
    return df_pen[mask].copy()

def display_case_summary(row, penalties, label=''):
    row = dict(row or {})
    if row.get('inspection_key') not in (None, 'N/A'):
        row['inspection_key'] = mask_identifier(row.get('inspection_key'))
    print_business_case(row, penalties, label=label)

def select_examples_by_decision(df, decision, n=3):
    if df is None or len(df) == 0 or 'decision' not in df.columns:
        return pd.DataFrame()
    sub = df[df['decision'] == decision].copy()
    if len(sub) == 0:
        return sub
    if 'vhs_final_score' in sub.columns:
        sub = sub.sort_values('vhs_final_score').reset_index(drop=True)
    if len(sub) <= n:
        return sub
    step = (len(sub) - 1) / max(n - 1, 1)
    idxs = [int(round(i * step)) for i in range(n)]
    return sub.iloc[idxs]

def select_examples_by_status(df_pen, df_sc, status, n=5):
    status_col = 'normalized_status' if df_pen is not None and 'normalized_status' in df_pen.columns else 'observed_status'
    if df_pen is None or len(df_pen) == 0 or status_col not in df_pen.columns:
        return pd.DataFrame(), pd.DataFrame()
    keys = df_pen[df_pen[status_col] == status]['inspection_key'].unique()[:n]
    pen_sub = df_pen[df_pen['inspection_key'].isin(keys)].copy()
    sc_sub  = (df_sc[df_sc['inspection_key'].isin(keys)].copy() if df_sc is not None and len(df_sc) > 0 and 'inspection_key' in df_sc.columns else pd.DataFrame())
    return sc_sub, pen_sub

def select_ambiguous_examples(df_pen, df_sc):
    if df_pen is None or len(df_pen) == 0 or 'valeur_controle' not in df_pen.columns:
        return pd.DataFrame(), pd.DataFrame()
    ambig = df_pen['valeur_controle'].astype(str).str.lower().isin(['proposition faite', 'non'])
    keys  = df_pen[ambig]['inspection_key'].unique()
    pen_sub = df_pen[df_pen['inspection_key'].isin(keys)].copy()
    sc_sub  = (df_sc[df_sc['inspection_key'].isin(keys)].copy() if df_sc is not None and len(df_sc) > 0 and 'inspection_key' in df_sc.columns else pd.DataFrame())
    return sc_sub, pen_sub

print('[OK] Fonctions d\'exemples métier chargées.')
```
